```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\LOQ\\Documents\\PYTHON\\Library\\Scikit Learn\\crime_dataset_india.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df['Weapon Used'] = df['Weapon Used'].fillna('Unknown')
df = df.drop(['Date Case Closed', 'Police Deployed', 'Date Reported', 'Date Reported'],axis=1)
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df['Date of Occurrence'] = pd.to_datetime(df['Date of Occurrence'])
```

```python
df['Time of Occurrence'] = pd.to_datetime(df['Time of Occurrence'], dayfirst=True)
```

```python
df['Hour'] = df['Time of Occurrence'].dt.hour
df['Minute'] = df['Time of Occurrence'].dt.minute
df['TotalMinutes'] = df['Hour'] * 60 + df['Minute']
df['Time_sine'] = np.sin(2 * np.pi * df['TotalMinutes'] / 1440)
df['Time_cos'] = np.cos(2 * np.pi * df['TotalMinutes'] / 1440)
```

```python
df['Year'] = df['Date of Occurrence'].dt.year
df['Month'] = df['Date of Occurrence'].dt.month
df['Day'] = df['Date of Occurrence'].dt.day
df['DayOfWeek'] = df['Date of Occurrence'].dt.dayofweek
df['Month_sine'] = np.sin(2 * np.pi * df['Month'] / 12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)
df['DoW_sine'] = np.sin(2 * np.pi * df['DayOfWeek'] / 7)
df['DoW_cos'] = np.cos(2 * np.pi * df['DayOfWeek'] / 7)
```

```python
df.head()
```

```python
df = df.drop(['Date of Occurrence', 'Time of Occurrence', 'Hour', 'Minute', 'TotalMinutes', 'Year', 'Month', 'Day', 'DayOfWeek'], axis=1)
```

```python
df.head()
```

```python
df = df.drop(['Case Closed', 'Report Number'], axis=1)
```
